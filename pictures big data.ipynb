# Binomiale

```python
p = 0.2
for n in [10, ]: 
    x = range(int(n))
    xx = np.array(x, dtype=float)/n
    plt.plot(xx,stats.binom(n,p).pmf(x), "-o", label="Binomial law\np=0.2 N=10")
plt.legend()
```

```python
p = 0.2
for n in [10, 100, 1000, 1E6]: 
    x = range(int(n))
    xx = np.array(x, dtype=float)/n
    plt.plot(xx,np.sqrt(n)*stats.binom(n,p).pmf(x), label="n = %d"%int(n))
plt.legend()
```

```python
N = 10000
D = 1
x = np.random.rand(N,D)-0.5
dist = np.zeros(N)
for d in range(D):
    dist += x[:,d]**2
plt.hist(np.sqrt(dist)/D, range=[0,0.5], bins=100);
#plt.xticks([])
plt.yticks([])
plt.title('distance to center distribution D=%d'%D)
```

```python
N = 1000
D = 100
x = np.random.rand(N,D)-0.5
dist = np.zeros((N,N))
for d in range(D):
    xd = x[:,d].reshape(1,N)
    xxd = (xd-xd.T)**2
    dist += xxd
plt.hist(np.sqrt(dist.ravel()),bins=300);
#plt.xticks([])
plt.yticks([])
plt.title('inter point distance D=%d'%D)
```

```python
N = 1000
RM = np.random.randn(N,N)
fig, (ax1,ax2,) = plt.subplots(ncols=2, figsize=(10,3))
cm = 'gnuplot2'
im1 = ax1.imshow(RM, interpolation="None", cmap=cm)
ax1.set_title('$A$')
Q = np.dot(RM,RM.T)
im = ax2.imshow(Q, interpolation=None, cmap=cm)
ax2.set_title('$A A^t$')
plt.colorbar(im1)
plt.colorbar(im)
ax1.set_axis_off()
ax2.set_axis_off()
```
